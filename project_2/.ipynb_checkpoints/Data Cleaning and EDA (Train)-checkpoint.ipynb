```python
def Garage_Age(date):
    
    if date == 'NA':
        age = 0
        
    elif date < 1950:
        age = 1
        
    elif date < 1960:  
        age = 2
        
    elif date < 1970:
        age = 3
        
    elif date < 1980:
        age = 4
        
    elif date < 1990:
        age = 5
        
    elif date < 2000:
        age = 6
        
    else: 
        age = 7
        
    return age

#since there are some houses without garages, we change the Garage Yr Blt feature to a discrete feature
```

```python
housing['Garage Age'] = housing['Garage Yr Blt'].apply(Garage_Age)
housing['Garage Age'].value_counts()
```
